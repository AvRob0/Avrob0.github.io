```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exoplanet_data = pd.read_csv("/content/Exoplanet_2024_April_06.csv")
exoplanet_data.head()
exoplanet_data.columns
```

```python
for i in exoplanet_data.columns:
  if exoplanet_data[i].isna().sum() != 0:
    print('null values in: ', i, 'column', exoplanet_data[i].isna().sum())
```

```python
missing_data = exoplanet_data.isna().sum().sort_values(ascending=False)
missing_data[:10].plot(kind='bar')
plt.title('Missing Data in the Dataset')
plt.xlabel('Columns with Missing Values')
plt.ylabel('Count of Missing Values')
```
