```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
exoplanet_data = pd.read_csv("/content/Exoplanet_2024_April_06.csv")
exoplanet_data.head()


```

```python
# prompt: clean the first 10 rows of data

# Drop rows with missing values in the 'planet_name' column
exoplanet_data = exoplanet_data.dropna(subset=['planet_name'], how='any').reset_index(drop=True)

# Drop rows with missing values in the first 10 rows
exoplanet_data = exoplanet_data.dropna(how='any', axis=0).iloc[:10, :]

# Print the first 10 rows of the cleaned data
print(exoplanet_data.head(10))
```

```python
# prompt: create a DataFrame from exoplanet_data.head(10)

exoplanet_df = pd.DataFrame(exoplanet_data.head(10))
print(exoplanet_df)
```

```python
# prompt: from the DataFrame exoplanet_df describe it

print(exoplanet_df.describe())
```

```python
# prompt: show the relationship between the planets stellar_surface_gravity_log10_(cm/s**2)

plt.scatter(exoplanet_df['stellar_surface_gravity_log10_(cm/s**2)'], exoplanet_df['planet_name'])
plt.xlabel('Stellar Surface Gravity (log10(cm/s^2))')
plt.ylabel('Planet Name')
plt.title('Relationship between Stellar Surface Gravity Between Each Other And Earths')
plt.show()
```

```python
# prompt: show the relationship between the planets stellar_surface_gravity_log10_(cm/s**2) compared to earths

# Calculate Earth's stellar_surface_gravity_log10_(cm/s^2)
earth_gravity = 9.80665 # m/s^2
earth_gravity_log10 = np.log10(earth_gravity * 100) # convert to cm/s^2 and take log10

# Plot a horizontal line for Earth's gravity
plt.axhline(y=earth_gravity_log10, color='red', linestyle='--', label='Earth\'s Gravity')

# Show the plot again
plt.scatter(exoplanet_df['stellar_surface_gravity_log10_(cm/s**2)'], exoplanet_df['planet_name'])
plt.xlabel('Stellar Surface Gravity (log10(cm/s^2))')
plt.ylabel('Planet Name')
plt.title('Relationship between Stellar Surface Gravity Between Each Other And Earths')
plt.legend()
plt.show()
```

```python
# prompt: show the difference between TOI-1386 c planet_radius_(earth_radius) compared to earths

# Calculate Earth's radius in Earth radii
earth_radius = 6371 # km
earth_radius_earth_radii = earth_radius / 6371

# Calculate the difference between TOI-1386 c planet_radius_(earth_radius) and Earth's radius
toi_1386_c_earth_radius = 1.25 # Earth radii
difference = toi_1386_c_earth_radius - earth_radius_earth_radii

# Print the difference
print("The difference between TOI-1386 c planet_radius_(earth_radius) and Earth's radius is:", difference, "Earth radii")
```

```python
# prompt: show the visualization difference between TOI-1386 c planet_radius_(earth_radius) compared to earths

# Create a bar chart to visualize the difference
x = ['TOI-1386 c', 'Earth']
y = [toi_1386_c_earth_radius, earth_radius_earth_radii]

plt.bar(x, y)
plt.xlabel('Planet')
plt.ylabel('Radius (Earth radii)')
plt.title('Comparison of TOI-1386 c and Earth Radius')
plt.show()
```

```python
# prompt: The difference between TOI-1386 c planet_radius_(jupiter_radius) compared to earths

# Calculate Earth's radius in Jupiter radii
earth_radius_jupiter_radii = earth_radius / 71492

# Calculate the difference between TOI-1386 c planet_radius_(jupiter_radius) and Earth's radius
toi_1386_c_jupiter_radius = 0.121 # Jupiter radii
difference_jupiter = toi_1386_c_jupiter_radius - earth_radius_jupiter_radii

# Print the difference
print("The difference between TOI-1386 c planet_radius_(jupiter_radius) and Earth's radius is:", difference_jupiter, "Jupiter radii")
```

```python
# prompt: show a visualization of the difference between TOI-1386 c planet_radius_(jupiter_radius) compared to earths

import matplotlib.pyplot as plt

# Define data
data = {'Earth': earth_radius_jupiter_radii, 'TOI-1386 c': toi_1386_c_jupiter_radius}

# Create a bar chart
plt.bar(data.keys(), data.values())

# Add title and axis labels
plt.title('Comparison of TOI-1386 c and Earth Radii (Jupiter Radii)')
plt.xlabel('Planet')
plt.ylabel('Radius (Jupiter Radii)')

# Add a horizontal line for Earth's radius
plt.axhline(y=earth_radius_jupiter_radii, color='red', linestyle='--', label='Earth\'s Radius')

# Add legend and show plot
plt.legend()
plt.show()
```

```python
# prompt: show a visualization of the difference between TOI-1386 c orbital_period compared to earths

# Calculate Earth's orbital period in days
earth_orbital_period = 365.25

# Calculate the difference between TOI-1386 c orbital_period and Earth's orbital period
difference_orbital_period = exoplanet_df['orbital_period'].iloc[0] - earth_orbital_period

# Print the difference
print("The difference between TOI-1386 c orbital_period and Earth's orbital period is:", difference_orbital_period, "days")

# Create a bar chart to visualize the difference
x = ['TOI-1386 c', 'Earth']
y = [exoplanet_df['orbital_period'].iloc[0], earth_orbital_period]

plt.bar(x, y)
plt.xlabel('Planet')
plt.ylabel('Orbital Period (days)')
plt.title('Comparison of TOI-1386 c and Earth Orbital Period')
plt.show()
```
