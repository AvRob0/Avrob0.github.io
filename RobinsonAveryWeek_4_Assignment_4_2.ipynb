Download the dataset and load it as a pandas dataframe and perform the following:

1. Using pandas to insert a new record into the dataset with the following details:

Fast-Food Chain: "Subway",

Systemwide Sales: $1000 million,

Average Sales per Unit: $400 thousand,

Franchised Stores: 2000,

Company Stores: 500,

2021 Total Units: 2500,

Total Change in Units from 2020: 50

```python
import pandas as pd
from datetime import date
import numpy as np
# created the DataFrame for the new data
new_data = [["Subway", "$1000 million", "$400 thousand", 2000, 500, 2500, 50]]
columns = ["Fast-Food Chain", "Systemwide Sales", "Average Sales per Unit", "Franchised Stores", "Company Stores", "2021 Total Units", "Total Change in Units from 2020"]
df = pd.DataFrame(new_data, columns=columns)
df
```

```python
# convert the DataFrame to dictionary format
dict_row = df.to_dict('records')[0]
dict_row
```

```python
# get the existing data from the csv file and convert it into pandas DataFrame
existing_record = pd.read_csv('/content/Fast-Food Chains in USA.csv', encoding='utf-8', engine='python')
# Then append the new data to the existing data
existing_record = pd.concat([existing_record, pd.DataFrame([dict_row])], ignore_index=True)
# Write the updated DataFrame to the csv file
existing_record.to_csv('/content/Fast-Food Chains in USA.csv', index=False, encoding='utf-8', header=True)
```

3. Using pandas to calculate the average systemwide sales for each category of fast food chains, grouped by category.

```python
calculate_average = pd.read_csv('/content/Fast-Food Chains in USA.csv', encoding='utf-8', engine='python')
calculate_average
```

```python
grouped_df = calculate_average.groupby('Fast-Food Chains')['U.S. Systemwide Sales (Millions - U.S Dollars)']

average_sales = grouped_df.mean()
print(average_sales)
```

4.Using pandas retrieve the top 10 fast food chains with the highest average sales per unit, ordered by average sales per unit in descending order.
